```python
from itertools import combinations

def greedy_combinations(data):
    seen_combinations = set()
    remaining_data = {k: dict(v) for k, v in data.items()}  # Copy the original structure
    all_values_sorted = sorted(
        [(p, k, v) for p, values in data.items() for k, v in values.items()], key=lambda x: x[2]
    )  # Sort all values globally
    results = []

    print(all_values_sorted)

    while any(remaining_data.values()) or all_values_sorted:
        iteration_result = {1: [], 2: [], 3: []}

        # Extract minimum value for each p_i
        min_values = {}
        for key, values in remaining_data.items():
            if values:
                min_val_key = min(values, key=values.get)
                min_values[key] = min_val_key
                del remaining_data[key][min_val_key]  # Remove the selected minimum

        # Generate combinations
        min_list = list(min_values.values())
        iteration_result[1] = min_list

        for r in range(2, len(min_list) + 1):
            new_combos = {"-".join(sorted(combo)) for combo in combinations(min_list, r)}
            new_combos -= seen_combinations  # Remove already seen combinations
            if new_combos:  # Only store non-empty combinations
                iteration_result[r] = list(new_combos)
                seen_combinations.update(new_combos)

        results.append(iteration_result)

        # If any group is empty, refill it with the next available minimum globally
        empty_keys = [key for key in remaining_data if not remaining_data[key]]
        for key in empty_keys:
            found = False
            for i, (next_p, next_v, next_val) in enumerate(all_values_sorted):
                if next_p == key:  # Ensure only the correct empty group is refilled
                    remaining_data[key] = {next_v: next_val}
                    all_values_sorted.pop(i)  # Remove from sorted list
                    found = True
                    break  # Move to the next empty key
            if not found:
                del remaining_data[key]  # Remove key if no more values exist

    return results

# Example usage
data = {
    'p1': {'v1': 0.8, 'v2': 0.5, 'v3': 0.2},
    'p2': {'v4': 0.7, 'v5': 0.6},
    'p3': {'v6': 0.3, 'v7': 0.4, 'v8': 0.5}
}

results = greedy_combinations(data)
for i, res in enumerate(results, 1):
    print(f"Iteration {i}: {res}")
```

```python
import numpy as np

class TreeNode:
    def __init__(self, value, parent=None):
        self.value = value
        self.children = []
        self.parent = parent

    def add_child(self, child):
        self.children.append(child)


def relative_increase_heuristic(values, new_value, threshold=0.5):
    """
    Check if adding the new_value to the values results in an increase
    greater than the threshold percentage of the average of the current values.
    """
    if not values:
        return False  # No values yet, so no increase to measure.

    # Calculate the average of current values
    current_sum = sum(values)
    average = current_sum / len(values)

    # Calculate the potential new sum if the value is added
    new_sum = current_sum + new_value

    # Calculate the relative increase
    relative_increase = (new_sum - current_sum) / average

    # If the relative increase is greater than the threshold, return True (stop)
    return relative_increase > threshold

def dfs(node, path=[], values=[], sets=[]):
    if node is None:
        return

    new_path = path + [node.value] if node.value else path
    new_values = values + [sets[len(new_path) - 1][node.value]] if node.value else values

    # Compute variance-based constraint
    if len(new_values) > 1:
        new_value = new_values[-1]
        if relative_increase_heuristic(new_values[:-1], new_value, 1):
            return  # Stop further exploration from this path

    print(f"{', '.join(new_path)} (Sum: {sum(new_values):.2f}, Variance: {np.var(new_values) if len(new_values) > 1 else 0:.2f})")  # Print combination with sum and variance

    for child in node.children:
        dfs(child, new_path, new_values, sets)

# Generating the tree based on sorted sets
def generate_combinations(sets, parent=None):
    if not sets:
        return None

    current_set = sets[0]
    root = TreeNode(None) if parent is None else parent

    sorted_items = sorted(current_set.items(), key=lambda x: x[1])  # Sort by value

    for key, _ in sorted_items:
        child = TreeNode(key, root)
        root.add_child(child)
        generate_combinations(sets[1:], child)

    return root

# Define the sets with values
sets_list = [
    {"v1": 0.2, "v2": 0.3, "v3": 0.8},
    {"v4": 0.5, "v5": 0.7},
    {"v6": 0.6, "v7": 0.7, "v8": 0.8}
]

# Execute DFS for each subset of sets
for i in range(len(sets_list)):
    subset = sets_list[i:]
    root = generate_combinations(subset)
    dfs(root, sets=subset)
```

```python
class TreeNode:
    def __init__(self, value, parent=None):
        self.value = value
        self.children = []
        self.parent = parent

    def add_child(self, child):
        self.children.append(child)

def dfs(node, path=[], values=[], sets=[]):
    if node is None:
        return

    new_path = path + [node.value] if node.value else path

    if node.value:
        for i, s in enumerate(sets):
            if node.value in s:
                new_values = values + [s[node.value]]
                break
        else:
            return  # Skip invalid nodes
    else:
        new_values = values

    print(f"{', '.join(new_path)} (Sum: {sum(new_values):.2f})")  # Print combination with sum

    for child in node.children:
        dfs(child, new_path, new_values, sets)

# Generating the tree based on sorted sets
def generate_combinations(sets, parent=None):
    if not sets:
        return None

    root = TreeNode(None) if parent is None else parent

    for i, current_set in enumerate(sets):
        sorted_items = sorted(current_set.items(), key=lambda x: x[1])  # Sort by value

        for key, _ in sorted_items:
            child = TreeNode(key, root)
            root.add_child(child)
            generate_combinations(sets[i+1:], child)  # Ensure all possible combinations are explored

    return root

# Define the sets with values
sets_list = [
    {"age-1": 0.5, "age-0": 0.37},
    {"race-1": 0.21, "race-0": 0.46},
    {"sex-1": 0.41, "sex-0": 0.25}
]

# Execute DFS for each subset of sets
for i in range(len(sets_list)):
    subset = sets_list[i:]
    root = generate_combinations(subset)
    dfs(root, sets=subset)
```
