# Profiling and running an ML algorithm on the US Census Dataset

```python
#Importing required packages.
import pandas as pd

%matplotlib inline
```

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data"

df = pd.read_table(filepath_or_buffer=url, names = ['status', 'month', 'credit-history',
'purpose', 'credit-amount', 'savings', 'employment', 'investment-as-income-percentage', 'sex-status',
 'other-debtors', 'residence-since', 'property', 'age', 'installment-plans', 'housing',
 'number-of-credits',  'skill-level', 'people-liable-for', 'telephone', 'foreign-worker',
 'credit'
], sep=" ")

df.reset_index(drop=True, inplace=True)

df.head()
```

```python
df['credit'] = df['credit'].replace(1, 'Good Credit')
df['credit'] = df['credit'].replace(2, 'Bad Credit')
```

```python
df['sex'] = df['sex-status'].replace({'A91': 'Male', 'A93': 'Male', 'A94': 'Male'})
df['sex'] = df['sex'].replace({'A92': 'Female', 'A95': 'Female'})

df.drop(['sex-status'], axis=1, inplace=True)
```

```python
df.head()
```

```python
bins = [0, 26, 120]
labels = ['Young', 'Old']

df['age'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
df.head()
```

```python
df.to_csv('german.csv', index=False)
```
