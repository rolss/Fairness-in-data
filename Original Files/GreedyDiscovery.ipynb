### Threshold: length of path explored against sum of values in the path

* Make a tree structure with every combination of attributes (multiple paths). Holds the form:
```
Root
├── v1 (from set0)
│   ├── v4 (from set1)  
│   │   ├── v6 (from set2)
│   │   ├── v7 (from set2) 
│   │   └── v8 (from set2)
│   └── v5 (from set1)
│       ├── v6 (from set2)
│       ├── v7 (from set2)
│       └── v8 (from set2)
├── v2 (from set0)
│   └── ... (same pattern)
└── v3 (from set0)
    └── ... (same pattern)
```

* DFS exploration of the tree, exploring only nodes where the sum of the path $\leq$ threshold, and pruning the node otherwise (not exploring that subtree). 
* Explores most promising combinations first (sorted)



````python
class TreeNode:
    def __init__(self, value, parent=None):
        self.value = value
        self.children = []
        self.parent = parent

    def add_child(self, child):
        self.children.append(child)

def dfs(node, path=[], values=[], sets=[]):
    if node is None:
        return

    new_path = path + [node.value] if node.value else path
    new_values = values + [sets[len(new_path) - 1][node.value]] if node.value else values

    # Compute the maximum allowed sum based on the number of items
    max_sum = 0.5 * len(new_values)
    total_sum = sum(new_values)

    # Check if the combination exceeds the allowed sum
    if total_sum > max_sum:
        return  # Stop further exploration from this path

    print(f"{', '.join(new_path)} (Sum: {total_sum:.2f}) (Max: {max_sum:.2f})")  # Print combination with sum
    for child in node.children:
        dfs(child, new_path, new_values, sets)

# Generating the tree based on sorted sets
def generate_combinations(sets, parent=None):
    if not sets:
        return None

    current_set = sets[0] # Select first tuple
    root = TreeNode(None) if parent is None else parent

    sorted_items = sorted(current_set.items(), key=lambda x: x[1])  # Sort by value

    for key, _ in sorted_items:
        child = TreeNode(key, root)
        root.add_child(child)
        generate_combinations(sets[1:], child)

    return root

# Define the sets with values
sets_list = [
    {"v1": 0.2, "v2": 0.3, "v3": 0.8},
    {"v4": 0.5, "v5": 0.7},
    {"v6": 0.6, "v7": 0.7, "v8": 0.8}
]

# Execute DFS for each subset of sets
for i in range(len(sets_list)):
    subset = sets_list[i:]
    root = generate_combinations(subset) # Generate tree
    dfs(root, sets=subset) # Explore tree
````

### Threshold: Relative increase when adding new value 

If the relative increase is > threshold, prune. 

````python
import numpy as np

class TreeNode:
    def __init__(self, value, parent=None):
        self.value = value
        self.children = []
        self.parent = parent

    def add_child(self, child):
        self.children.append(child)


def relative_increase_heuristic(values, new_value, threshold=0.5):
    """
    Check if adding the new_value to the values results in an increase
    greater than the threshold percentage of the average of the current values.
    """
    if not values:
        return False  # No values yet, so no increase to measure.

    # Calculate the average of current values
    current_sum = sum(values)
    average = current_sum / len(values)

    # Calculate the potential new sum if the value is added
    new_sum = current_sum + new_value

    # Calculate the relative increase
    relative_increase = (new_sum - current_sum) / average

    # If the relative increase is greater than the threshold, return True (stop)
    return relative_increase > threshold

def dfs(node, path=[], values=[], sets=[]):
    if node is None:
        return

    new_path = path + [node.value] if node.value else path
    new_values = values + [sets[len(new_path) - 1][node.value]] if node.value else values

    # Compute variance-based constraint
    if len(new_values) > 1:
        new_value = new_values[-1]
        if relative_increase_heuristic(new_values[:-1], new_value, 1):
            return  # Stop further exploration from this path

    print(f"{', '.join(new_path)} (Sum: {sum(new_values):.2f}, Variance: {np.var(new_values) if len(new_values) > 1 else 0:.2f})")  # Print combination with sum and variance

    for child in node.children:
        dfs(child, new_path, new_values, sets)

# Generating the tree based on sorted sets
def generate_combinations(sets, parent=None):
    if not sets:
        return None

    current_set = sets[0]
    root = TreeNode(None) if parent is None else parent

    sorted_items = sorted(current_set.items(), key=lambda x: x[1])  # Sort by value

    for key, _ in sorted_items:
        child = TreeNode(key, root)
        root.add_child(child)
        generate_combinations(sets[1:], child)

    return root

# Define the sets with values
sets_list = [
    {"v1": 0.2, "v2": 0.3, "v3": 0.8},
    {"v4": 0.5, "v5": 0.7},
    {"v6": 0.6, "v7": 0.7, "v8": 0.8}
]

# Execute DFS for each subset of sets
for i in range(len(sets_list)):
    subset = sets_list[i:]
    root = generate_combinations(subset)
    dfs(root, sets=subset)
````
